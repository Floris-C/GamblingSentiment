```python
import snscrape.modules.twitter as sntwitter
from nltk.sentiment.vader import SentimentIntensityAnalyzer

vader = SentimentIntensityAnalyzer()

dct = {"topic": [], "compound": []}
topics = ["#lec", "happy", "angry", "mid", "gambl"]

for topic in topics:
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(topic).get_items()):
        if i > 500: break
        dct["topic"].append(topic)
        dct["compound"].append(vader.polarity_scores(tweet.rawContent)["compound"])
reference_df = pd.DataFrame(dct)
```

```python
for i,tweet in enumerate(sntwitter.TwitterSearchScraper("gamble OR gambling").get_items()):
        if i > 500: break
        dct["topic"].append("gamble OR gambling")
        dct["compound"].append(vader.polarity_scores(tweet.rawContent)["compound"])
reference_df = pd.DataFrame(dct)
```
